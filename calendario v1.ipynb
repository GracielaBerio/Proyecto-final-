# Calendario construcción

## Objetivo
Pintar el calendario de un color distinto según la prioridad de cada tarea (Alta:rojo / Media:amarillo / Baja:verde).  
Cada evento se asocia con un color en el calendario.

## Limpiar eventos previos
- Borra todos los eventos anteriores del calendario.
  
```python
try:
    self.calendar.calevent_remove('all')
except Exception:
    pass
```

## Definir colores por prioridad
- Se asigna un color a cada prioridad usando códigos hexadecimales.
  
```python
colors = {
    "Alta": "#ff4d4d",    # rojo
    "Media": "#ffe066",   # amarillo
    "Baja": "#66ff99"     # verde
}
```
## Crear eventos por tarea
- PErmite recorre todas las tareas ingresadas, de acuerdo a su: id.. titulo, descripción, fecha y prioridad, si la tarea tiene fecha y prioridad válida, crea un evento en el calendario con su color correspondiente.
  
```python
for t in get_tasks():
    due = t[3]
    pr = t[4]
    if due and pr in colors:
        try:
            dt = datetime.setiembre(due, "%Y-%m-%d")
            self.calendar.calevent_create(dt, t[1], pr)
        except Exception:
            continue
```
## Aplicar colores según prioridad
 Prioridad(`Alta`, `Media`, `Baja`).
 
```python
for pr, col in colors.items():
    try:
        self.calendar.tag_config(pr, background=col, foreground="black")
    except Exception:
        pass
```

## Función on_calendar_selecc
- Cuando el usuario selecciona una fecha, se obtiene esa fecha y se filtran las tareas que coinciden.
  
```python
def on_calendar_select(self, event):
    if not self.calendar:
        return
    selected_date = self.calendar.get_date()
    self.refresh_tasks(filter_date=selected_date)
```

## Función show_all_tasks
Quita el filtro y muestra todas las tareas nuevamente.

```python
def show_all_tasks(self):
    self.refresh_tasks(filter_date=None)
```
